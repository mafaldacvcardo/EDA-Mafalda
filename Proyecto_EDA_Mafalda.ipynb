# Movilizaciones de extrema-derecha

# Tema

La creación del discurso del odio se considera el renacimiento de un cuerpo autoritario: desde los movimientos de extrema derecha
hasta los partidos políticos. Se ponen en juego ideas como el rechazo a la inmigración y el no reconocimiento de las minorías,
dibujando una polarización social que separa a la multitud del peligro exterior. La política de extrema derecha utiliza activamente
el espacio público para hacer circular su discurso y poder fomentando mecanismos de control vigilante y marginación.
Analizar el comportamiento de protesta significa apelar a una formación diferente de poder-en-la-masa.

La política de extrema derecha es un buen ejemplo de ello a través de su práctica de despolitizar, demonizar y neutralizar a ciertos grupos
para convertirlos en personas sin subjetividad política. Las concentraciones, las marchas y los símbolos nacionalistas,
contribuyen a activar a los simpatizantes, mediante marcos de exclusión y superioridad, ya que el nacionalismo rechaza el multiculturalismo
replegándose en una negación ideológica simplificada.

La acción colectiva de las contramanifestaciones ofrece una fuerza transformadora, recuperando la pluralidad, la interdependencia y
la humanidad compartida. La resistencia multitudinaria, arraigada en la vulnerabilidad y la unión, desbarata las ideologías de extrema derecha,
desafiando su control homogeneizador y reimaginando posibilidades democráticas que priorizan la inclusión y la igualdad sobre la exclusión
y la dominación.

# Hipótesis

Las hipótesis elegidas para guiar este proyecto pretenden comprobar cómo el discurso material/redistributivo es secundario frente al identitario:

 -	Salvo en momentos puntuales de crisis (por ejemplo, entre 2012 y 2014), los temas relacionados con el bienestar (“welfare”)
    no ocupan un lugar central en la agenda de movilización de la extrema derecha.
    
 -	La extrema derecha recurre a los temas de bienestar principalmente cuando pueden integrarse en un marco nacionalista o excluyente (“welfare         chauvinism”).
 
 -	La narrativa económica de la extrema derecha es reactiva (ante crisis o competencia política), no un componente estable de su agenda.
 
 -	Los temas relacionados con Islam, inmigración e identidad se consolidan como pilares del discurso de la extrema derecha,
    especialmente a partir de 2015.

- Las principales causas de las protestas (como la inmigración, el islam y la seguridad) varían entre los distintos países.

- Los niveles e intensidades de protesta más elevados se asocian con movilizaciones en los que participan varios actores.

- Las contramanifestaciones tienden a tener más actores participantes que las movilizaciones sin oposición.

- La frecuencia de las contramanifestaciones aumenta con el tiempo (por ejemplo, después de 2015) a medida que crece la polarización.
